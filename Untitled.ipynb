```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Завантаження даних
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'FB']
data = yf.download(tickers, start='2020-01-01', end='2023-01-01')['Adj Close']

# Розрахунок середньої доходності та ковариацій
returns = data.pct_change().dropna()
mean_returns = returns.mean()
cov_matrix = returns.cov()

# Функція для розрахунку ризику портфелю
def portfolio_volatility(weights, mean_returns, cov_matrix):
    return np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))

# Обмеження та обмеження для оптимізації
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
bounds = tuple((0, 1) for _ in range(len(tickers)))

# Початкові ваги
init_guess = len(tickers) * [1. / len(tickers)]

# Оптимізація портфелю
opt_result = minimize(portfolio_volatility, init_guess, args=(mean_returns, cov_matrix), 
                      method='SLSQP', bounds=bounds, constraints=constraints)

# Оптимальні ваги
opt_weights = opt_result.x

# Візуалізація результатів
plt.figure(figsize=(10, 6))
plt.pie(opt_weights, labels=tickers, autopct='%1.1f%%', startangle=140)
plt.title('Оптимальний розподіл активів')
plt.show()
```
